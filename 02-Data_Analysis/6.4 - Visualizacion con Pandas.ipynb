```python
import warnings
warnings.simplefilter('ignore')

import pandas as pd
import numpy as np

import pylab as plt
import seaborn as sns

%matplotlib inline
```

**Line plot**

```python
df=pd.Series(np.random.randn(1000), 
             index=pd.date_range('1/1/2000', periods=1000))

display(df.head())

df=df.cumsum()

df.plot();
```

```python
df=pd.DataFrame(np.random.randn(1000, 4), 
             index=pd.date_range('1/1/2000', periods=1000),
             columns=['a', 'b', 'c', 'd'])

df['letras']=['hola' for i in range(1000)]

display(df.head())

df=df.cumsum()

df.plot();
```

```python
df.a.plot();
```

```python
df.plot(x='a', y='b');
```

**Bar plot**

```python
df=df._get_numeric_data()

df.iloc[5].plot(kind='bar');
```

```python
df.iloc[5].plot(kind='bar')

plt.axhline(-3, color='k');
```

```python
df.index=[i for i in range(1000)]
```

```python
df.plot.bar();
```

```python
df.plot.bar(stacked=True);
```

```python
df.plot.barh(stacked=True);
```

**Histograma**

```python
df.plot.hist();
```

```python
df.plot.hist(stacked=True, bins=30);
```

```python
#help(df.plot())
```

```python
df.a.plot.hist(orientation='horizontal');
```

```python
df.a.plot.hist(orientation='horizontal', cumulative=True);
```

**Box plots**

```python
df.b.plot.box();
```

```python
df.boxplot();
```

```python
color={'boxes': 'DarkGreen',
       'whiskers': 'DarkOrange',
       'medians': 'DarkBlue',
       'caps': 'Gray'}

df.plot.box(color=color,
            sym='r+',
            vert=False
           );
```

**Area plot**

```python
df=df.apply(lambda x: abs(x))

df.plot.area();
```

```python
df.plot.area(stacked=False);
```

**Scatter plot**

```python
df.plot.scatter(x='c', y='a');
```

```python
ax_=df.plot.scatter(x='c', y='a')

df.plot.scatter(x='b', y='d', color='DarkGreen', ax=ax_);
```

```python
df.plot.scatter(x='c', y='a', c='b');
```

```python
df.plot.scatter(x='c', y='a', c='d', s=5);
```

```python
df.plot.scatter(x='c', y='a', c='d', s='b');
```
